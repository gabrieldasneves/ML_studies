Vamos classificar estrelas 

```
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as  pd
import os

# importando os dados
data= pd.read_csv("https://raw.githubusercontent.com/gabrieldasneves/ML_studies/master/star_classification/datasets_6_class.csv")
print(data.head())
```

```
# conhecendo os dados
print(data.shape)
print(data.info())
```

```
sns.pairplot(data = data, hue = "Star type")
plt.show()
```

```
# vamos ver um exemplo de distribuio
sns.distplot(data['Luminosity(L/Lo)'])
plt.show()
```

```
ax = sns.countplot(data["Star color"])
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

Pre processamento


```

data["Star color"] = data["Star color"].apply(lambda x: x.lower())
data["Star color"] = data["Star color"].apply(lambda x: x.replace("-"," "))
data["Star color"] = data["Star color"].apply(lambda x: x.replace(" ",""))
```
